# Data Analysis And Visualizations

```python
cancelled_perc=df['is_canceled'].value_counts(normalize = True)*100
# Get the value counts of 'is_canceled' and store the counts
cancel_counts = df['is_canceled'].value_counts()

# Prepare the x-labels (categories) and corresponding heights (values)
categories = ['Not Canceled', 'Canceled']
values = cancel_counts.values

# Create the plot
plt.figure(figsize=(5, 2))
plt.title("Reservation Status Count")

# Plot the bar chart
plt.bar(categories, values, edgecolor='k', width=0.7)

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(8, 4))
ax1=sbn.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation In Different Hotels", size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel'] =='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
city_hotel = df[df['hotel'] =='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title("Average Daily Rate in City and Resort Hotel", fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
# Convert the 'reservation_status_date' column to datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], errors='coerce')

# Now extract the month
df['month'] = df['reservation_status_date'].dt.month

# Plot the data
plt.figure(figsize=(16, 8))
ax1 = sbn.countplot(x='month', hue='is_canceled', data=df, palette='bright')
legend_label, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status Per Month', size=16)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')

# This line is incorrect, replace it with the correct one
# plt.legend('canceled', 'not cancelled') -> This is not how legend labels work

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=30)

# Add a palette to differentiate bar colors
sbn.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(), palette='Set2')

plt.show()
```

```python
# Get top 10 countries by cancellation
cancelled_data = df[df['is_canceled'] == 1]
top_10_countries = cancelled_data['country'].value_counts().head(10)  # Use .head(10) to get top 10

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.title("Top 10 Countries With Reservation Cancellation")
plt.pie(top_10_countries, autopct='%.2f', labels=top_10_countries.index)  # Use labels for country names

plt.show()
```

```python
canceled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date', inplace=True)

not_canceled_df_adr=df[df['is_canceled'] == 0]
not_canceled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date', inplace=True)


plt.figure(figsize=(20,6))
plt.title("Average Daily Rate")
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'], label='not canceled')
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label='canceled')
plt.legend(fontsize=27)
```

```python
canceled_df_adr = canceled_df_adr[(canceled_df_adr['reservation_status_date'] > '2016') & (canceled_df_adr['reservation_status_date'] < '2017-09')]
not_canceled_df_adr = not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date'] > '2016') & (not_canceled_df_adr['reservation_status_date'] < '2017-09')]
```

```python
# Convert 'reservation_status_data' to datetime if it's not already
canceled_df_adr['reservation_status_date'] = pd.to_datetime(canceled_df_adr['reservation_status_date'], errors='coerce')
not_canceled_df_adr['reservation_status_date'] = pd.to_datetime(not_canceled_df_adr['reservation_status_date'], errors='coerce')

# Plot the data
plt.figure(figsize=(20, 8))
plt.title("Average Daily Rate", fontsize=24)

# Plot for not cancelled
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'], label='Not Cancelled', color='blue')

# Plot for cancelled
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label='Cancelled', color='red')

# Add legend and formatting
plt.legend(fontsize=16, loc='upper right')
plt.xlabel('Reservation Date', fontsize=18)
plt.ylabel('Average Daily Rate (ADR)', fontsize=18)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)

plt.show()
```
